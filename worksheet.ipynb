```python
df_aggregated = df_agg.groupby(['county_code', 'Urban', 'State', 'County'], as_index=False).sum()
df_aggregated
```

```python
def calculate(row):
    if row['TractHispanic'] == 0:
        return None
    elif row["Urban"] == 1:
        return row['lahisphalf'] / row['TractHispanic']
    else:
        return row['lahisp10'] / row['TractHispanic']

def calculatewhite(row):
    if row['TractWhite'] == 0:
        return None
    elif row["Urban"] == 1:
        return row['lawhitehalf'] / row['TractWhite']
    else:
        return row['lawhite10'] / row['TractWhite']

def calculateblack(row):
    if row['TractBlack'] == 0:
        return None
    elif row["Urban"] == 1:
        return row['lablackhalf'] / row['TractBlack']
    else:
        return row['lablack10'] / row['TractBlack']  

def calculateblack(row):
    if row['TractAsian'] == 0:
        return None
    elif row["Urban"] == 1:
        return row['lablackhalf'] / row['TractBlack']
    else:
        return row['lablack10'] / row['TractBlack']
    
def calculateall(row):
    if row['Pop2010'] == 0:
        return None
    elif row["Urban"] == 1:
        return row['lapophalf'] / row['Pop2010']
    else:
        return row['lapop10'] / row['Pop2010']
    



# df['lahisp_5_10_ratio'] = df.apply(lambda row: row['lahisphalf'] / row['TractHispanic'] if row["Urban"] == 1 elif  row['lahisp10'] / row['TractHispanic'], axis=1)
df_aggregated['lahisp_5_10_ratio'] = df_aggregated.apply(calculate, axis=1)
df_aggregated['lawhite_5_10_ratio'] = df_aggregated.apply(calculatewhite, axis=1)
df_aggregated['lablack_5_10_ratio'] = df_aggregated.apply(calculateblack, axis=1)
df_aggregated['lapop_5_10_ratio'] = df_aggregated.apply(calculateall, axis=1)
```

```python
df_intermediate = df_aggregated[df_aggregated['Urban'] == 1][['county_code','lapop_5_10_ratio']].set_index('county_code').subtract(df_aggregated[df_aggregated['Urban'] == 0][['county_code', 'lapop_5_10_ratio']].set_index('county_code'))
df_intermediate = df_intermediate.rename(columns={'lapop_5_10_ratio': 'lapop_5_10_diff'})
df_intermediate
```

```python
df_intermediate.index
```

```python
df_aggregated = df_aggregated.join(df_intermediate, how='outer', on='county_code')
df_aggregated
```

```python
df_aggregated = df_aggregated[['county_code', 'Urban', 'State', 'County', 'Pop2010', 'TractWhite', 'TractBlack', 'TractHispanic', 'lawhite_5_10_ratio', 'lablack_5_10_ratio', 'lahisp_5_10_ratio', 'lapop_5_10_ratio', 'lapop_5_10_diff']]
display(df_aggregated)
```
